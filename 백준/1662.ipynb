# 코드
- 21.11 풀이
- 메모리 초과

```python
string = input().strip()
stack = []
for s in string:
    if s == ')': 
        temp = ''
        n = -1
        while stack and stack[n] != '(':
            temp = stack.pop() + temp
        stack.pop()
        stack.append(temp * int(stack.pop()))
    else:
        stack.append(s)
print(len(''.join(stack)))
```

# 코드
- 21.11 풀이
- 메모리 문제를 해결하기 위해 stack에 문자열 길이 저장

```python
# 29200	KB, 68 ms
string = input().strip()
stack = []
for s in string:
    print(stack)
    if s == ')': 
        temp = 0
        n = -1
        while stack and stack[n] != '(':
            x = stack.pop()
            if type(x) == int: temp += x
            else : temp += 1
        stack.pop()
        stack.append(temp * int(stack.pop()))
    else:
        stack.append(s)

answer = 0
for i in stack:
    if type(i) == int: answer += i
    else: answer += 1
print(answer)
```
