```python
# 136488 KB, 1120 ms (pypy3)
from itertools import accumulate
import sys
input = sys.stdin.readline

def BOJ_11066(N, arr):
    S = [0] + list(accumulate(arr))
    dp = [[0]*(N+1) for _ in range(N+1)]

    # dp[i][j] = i ~ j 까지 최소합
    # dp[i][j] = min(dp[i][k] + dp[k+1][j]) + sum(a[i] ~ a[j])
    for l in range(2, N+1): # 거리
        for x in range(1, N+2 - l): # 시작점 ==> 종료점 : x + l - 1
            dp[x][x+l-1] = min([dp[x][k] + dp[k+1][x+l-1] for k in range(x,x+l-1)]) + S[x+l-1] - S[x-1]

    # 시작점 ~ 종료점으로 for문 사용하면 dp가 제대로 채워지지 않음. 
    # 길이가 긴 dp의 경우에는 거리가 짧은 dp가 모두 채워져 있어야 정상적으로 코드가 수행됨
    # for i in range(1, N+1):
    #     for j in range(i+1, N+2):
    #         dp[i][j] = min([dp[i][k] + dp[k+1][j] for k in range(i+1, j)]) + S[j] - S[i-1]

    return dp[1][N]

T = int(input())
for _ in range(T):
    N = int(input())
    arr = list(map(int, input().split()))
    print(BOJ_11066(N, arr))
```

```python
# import sys
# input = sys.stdin.readline

def BOJ_11066(arr, n):
    S = [0]
    for j in arr:
        S.append(j + S[-1])

    # dp[a][b] = b ~ b+a-1 까지의 최소합
    dp = [[], [0] * n, [S[i+2] - S[i] for i in range(n-1)]]

    knuth_old = [1]*(n-1)
    for k in range(3, n+1):
        a, knuth_new = [], []
        for i in range(n-k+1):
            temp, temp2 = float('inf'), knuth_old[i]
            for m in range(knuth_old[i], knuth_old[i+1]+2):
                if temp >= (q := dp[m][i] + dp[k-m][m+i]):
                    temp, temp2 = q, m
            a.append(temp + S[k+i] - S[i]), knuth_new.append(temp2)
        dp.append(a)
        knuth_old = knuth_new
    return dp[-1][-1]

T = int(input())
for _ in range(T):
    N = int(input())
    arr = list(map(int, input().split()))
    print(BOJ_11066(arr, N))
```
